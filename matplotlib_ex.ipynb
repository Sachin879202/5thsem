```python
import pandas as pd
df = pd.read_excel('dataset/SaleData.xlsx')
df.info()
```

```python
df.dropna(inplace = True)
df.info()
```

```python
df
```

```python
df1 = df.groupby(df['Region']).sum()['Sale_amt']
df1
```

```python
import matplotlib.pyplot as plt
df1 = df.groupby(df['Region']).sum()['Sale_amt']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('sales amount as region',fontsize=15)    
ax.set_xlabel('Region',fontsize=15)    
ax.set_ylabel('sales amount',fontsize=15) 
ax.bar(df1.index,df1)
```

```python
df2 = df.groupby(df['Item']).sum()['Units']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('total units tally',fontsize=15)    
ax.set_xlabel('Items',fontsize=15)    
ax.set_ylabel('Units',fontsize=15) 
ax.bar(df2.index,df2)
```

```python
df3 = df.groupby(df['Manager']).sum()['Sale_amt']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('total sales by manager',fontsize=15)    
ax.set_xlabel('Manager',fontsize=15)    
ax.set_ylabel('sale amount',fontsize=15) 
ax.bar(df3.index,df3)
```
